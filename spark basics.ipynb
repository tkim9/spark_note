```
# adding new column -> df.withColumn('name', data)

df.withColumn('newage', df['age']*2).show()
```

```
# Change above is not permanent unless you make new variable

df.show()
```

```
# Renaming the column name    df.withColumnRenamed(old name, new name)

df.withColumnRenamed('age', 'my_new_age').show()
```

```
# Registering it to sql temporary view

df.createOrReplaceTempView('people')   # people is the new table name
```
